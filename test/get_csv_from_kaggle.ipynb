```python
from kaggle.api.kaggle_api_extended import KaggleApi
import os
os.environ['KAGGLE_CONFIG_DIR'] = '../.kaggle/'

api = KaggleApi()
api.authenticate()
dataset = "josephcheng123456/olympic-historical-dataset-from-olympediaorg"
api.dataset_download_files(dataset, path="../data/olympic", unzip=True)
```

To get the API Key from Kaggle : 
1. Create a folder named '.kaggle' in the dataeng_project
2. Home -> Settings -> Scroll down to the section API -> Create New Token
3. Save the kaggle.json file given inside the folder '.kaggle'
